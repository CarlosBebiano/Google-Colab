```python
sns.boxplot(salario['salary_in_usd']) #mostra em um grafico de candle de uma forma mais precisa onde se encontra o outlier
```

```python
# define os outliers atraves do metodo do z-score
upper_limit = salario['salary_in_usd'].mean() + 3*salario['salary_in_usd'].std() #calcula a oscilação da curva da Distribuição Normal
lower_limit = salario['salary_in_usd'].mean() - 3*salario['salary_in_usd'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
salario.loc[(salario['salary_in_usd'] > upper_limit) | (salario['salary_in_usd'] < lower_limit)] #verifica todos os valores para ver se entra na condição de ser um outliers
```

```python
# trimming - deleta todos os outliers encontrados na tabela e armazena em uma nova tabela
new_df = salario.loc[(salario['salary_in_usd'] <= upper_limit) & (salario['salary_in_usd'] >= lower_limit)]
print('before removing outliers:', len(salario))
print('after removing outliers:',len(new_df))
print('outliers:', len(salario)-len(new_df))
```

```python
sns.boxplot(new_df['salary_in_usd']) #mostra como ficou a tabela após deletar os outliers
```

```python
new_df[(salario['salary_in_usd'] <= upper_limit) & (salario['salary_in_usd'] >= lower_limit)] #mostra o menor,media e o maior salario pago
print("Minimum salary in USD: {}".format(new_df.salary_in_usd.min()))
print("Median  salary in USD: {}".format(new_df.salary_in_usd.median()))
print("Maximum salary in USD: {}".format(new_df.salary_in_usd.max()))
#sns.distplot(new_df["salary_in_usd"])
```
